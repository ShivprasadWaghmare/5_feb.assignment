It defines a set of attributes (data members)         and methods (functions) that the objects created from the class will have. class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model
        self.speed = 0

    def brake(self):
        self.speed -= 5

    def display_info(self):
        print(f"{self.make} {self.model} - Speed: {self.speed} km/h")

In this example, "Car" is the class with attributes like "make," "model," and "speed," and methods like "accelerate," "brake," and "display_info." The objects car1 and car2 are instances of the "Car" class, each having its own set of attributes and interacting with the defined methods.

5. What is inheritance? Give an example for each type of inheritance.
ans:
Inheritance is a concept in object-oriented programming where a new class (subclass or derived class) can inherit attributes and behaviors from an existing class (base class or parent class). This helps in reusing code and creating a hierarchy of classes.

1. **Single Inheritance:**
   - Example: 
     ```python
     class Animal:
         def speak(self):
             print("Animal speaks")

     class Dog(Animal):
         def bark(self):
             print("Dog barks")

     # Dog inherits from Animal, so it can use the speak method.
     my_dog = Dog()
     my_dog.speak()  # Output: Animal speaks
     ```

2. **Multiple Inheritance:**
   - Example:
     ```python
     class Flyable:
         def fly(self):
             print("Can fly")

     class Swimmable:
         def swim(self):
             print("Can swim")

     class Amphibian(Flyable, Swimmable):
         pass

     # Amphibian inherits from both Flyable and Swimmable.
     amphibian = Amphibian()
     amphibian.fly()  # Output: Can fly
     amphibian.swim() # Output: Can swim
     ```

3. **Multilevel Inheritance:**
   - Example:
     ```python
     class Vehicle:
         def start(self):
             print("Vehicle started")

     class Car(Vehicle):
         def drive(self):
             print("Car is being driven")

     class SportsCar(Car):
         def race(self):
             print("Sports car is racing")

     # SportsCar inherits from both Car and Vehicle.
     my_sports_car = SportsCar()
     my_sports_car.start()  # Output: Vehicle started
     my_sports_car.drive()  # Output: Car is being driven
     my_sports_car.race()   # Output: Sports car is racing
     ```

In each example:
- The base class provides common functionalities.
- The derived class inherits from the base class and may add new attributes or methods.
- Objects of the derived class can use both the inherited and newly defined functionalities.
